```python
from simpletransformers.language_modeling import LanguageModelingModel
from simpletransformers.language_generation import LanguageGenerationModel
```

## Initialization

```python
model_args = {
    'dataset_type': 'simple',
    'min_frequency': 5,
    'mlm': False,
    'num_train_epochs': 1,

    'output_dir': 'LM_outputs',
    'overwrite_output_dir': True,
}
```

```python
train_path = 'train_poems.txt'
```

```python
model = LanguageModelingModel(
    'gpt2',
    'sberbank-ai/rugpt3small_based_on_gpt2',
    use_cuda=True,
    args=model_args,
)
```

## Training

```python
model.train_model(train_path)
```

## Evaluation

```python
gen_model_args = {
    'do_sample': True,
    'max_length': 50,
    'verbose': True,
    'top_k': 50,
    'top_p': 0.95,
}
```

```python
gen_model = LanguageGenerationModel(
    'gpt2',
    'LM_outputs',
    use_cuda=True,
    args=gen_model_args,
)
```

```python
starts = [
    'Сижу я, значит, за столом, и вдруг...',
    'Чего?! Что ты несёшь? Ты ещё скажи, что',
]

gen_result = [gen_model.generate(text) for text in starts]
```

```python
print(*(line[0] for line in gen_result), sep='\n===\n')
```
